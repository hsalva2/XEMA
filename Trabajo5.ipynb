```python
# Cargar libraries usuales
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
```

```python
# Leer hoja de datos de salarios
datos = pd.read_csv('https://raw.githubusercontent.com/hsalva2/XEMA/main/df1.csv')
datos.head()
```

```python
reg1 = smf.ols(formula='Y ~ T',data=datos).fit()
print(reg1.summary2())
```

```python
reg2 = smf.ols(formula='Y ~ T + X1 + X2 + X3 + X4 ',data=datos).fit()
print(reg2.summary2())
```

```python
reg3 = smf.ols(formula='Y ~ T + X1 + X2  ',data=datos).fit()
print(reg3.summary2())
```

```python
reg4 = smf.ols(formula='Y ~ T + X1 + X2 + T:X1 + T:X2 ',data=datos).fit()
print(reg4.summary2())
```
